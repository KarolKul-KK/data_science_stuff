```python
from vectors import Vector, dot
```

```python
def sum_of_squares(v: Vector) -> float:
    '''Computing sum of elements v ** 2'''
    return do(v, v)
```

```python
from typing import Callable
```

```python
def difference_quotient(f: Callable[[float], float],
                       x: float,
                       h: float) -> float:
    return (f(x + h) - f(x)) / h
```

```python
def square(x: float) -> float:
    return x * x
```

```python
def derivative(x: float) -> float:
    return 2 * x
```

```python
xs = range(-10, 11)
actuals = [derivative(x) for x in xs]
estimates = [difference_quotient(square, x, h=0.001) for x in xs]
```

```python
import matplotlib.pyplot as plt
plt.title('Real deravatives and them estimated values')
plt.plot(xs, actuals, 'rx', label='Real values')
plt.plot(xs, estimates, 'b+', label='Estimated values')
plt.legend(loc=9)
plt.show()
```

```python
def partial_difference_quotient(f: Callable[[Vector], float],
                               v: Vector,
                               i: int,
                               h: float) -> float:
    w = [v_j + (h if j == i else 0)
        for j, v_j in enumerate(v)]
    
    return (f(w) - f(v)) / h
```

```python
def estimate_gradient(f: Callable[[Vector], float],
                     v: Vector,
                     h: float = 0.0001):
    return [partial_difference_quotient(f, v, i, h)
           for i in range(v)]
```

```python
import random
```

```python
from vectors import distance, add, scalar_multiply
```

```python
def gradient_step(v: Vector, gradient: Vector, step_size: float) -> Vector:
    assert len(v) == len(gradient)
    step = scalar_multiply(step_size, gradient)
    return add(v, step)
```

```python
def sum_of_squares_gradient(v: Vector) -> Vector:
    return [2 * v_i for v_i in v]
```

```python
v = [random.uniform(-10, 10) for i in range(3)]

for epoch in range(1000):
    grad = sum_of_squares_gradient(v)
    v = gradient_step(v, grad, -0.01)
    print(epoch, v)
    
assert distance(v, [0, 0, 0])
```

```python
inputs = [(x, 20 * x + 5) for x in range(-50, 50)]
```

```python
inputs
```

```python
def linear_gradient(x: float, y: float, theta: Vector) -> Vector:
    slope, intercept = theta
    predicted = slope * x + intercept
    error = (predicted - y)
    squared_error = error ** 2
    grad = [2 * error * x, 2 * error]
    return grad
```

```python
from vectors import vector_mean
```

```python
theta = [random.uniform(-1, 1), random.uniform(-1, 1)]

learning_rate = 0.001

for epoch in range(5000):
    # computing mean of gradient
    grad = vector_mean([linear_gradient(x, y, theta) for x, y in inputs])
    # step
    theta = gradient_step(theta, grad, -learning_rate)
    print(epoch, theta)
    
slope, intercept = theta
assert 19.9 < slope < 20.1, "slope value should be ~ 20"
assert 4.9 < intercept < 5.1, "intercept value should be ~ 5"
```

```python
from typing import TypeVar, List, Iterator
```

```python
T = TypeVar('T')

def minibatches(dataset: List[T],
               batch_size: int,
               shuffle: bool = True) -> Iterator[List[T]]:
    '''generate sample from dataset about batch size'''
    batch_starts = [start for start in range(0, len(dataset), batch_size)]
    
    if shuffle: random.shuffle(batch_starts)
    
    for start in batch_starts:
        end = start + batch_size
        yield dataset[start:end]
```

```python
theta = [random.uniform(-1, 1), random.uniform(-1, 1)]

for epoch in range(1000):
    for batch in minibatches(inputs, batch_size=20):
        grad = vector_mean([linear_gradient(x, y, theta) for x, y in batch])
        theta = gradient_step(theta, grad, -learning_rate)
    print(epoch, theta)
    
slope, intercept = theta
assert 19.9 < slope < 20.1, "slope value should be ~ 20"
assert 4.9 < intercept < 5.1, "intercept value should be ~ 5"
```
